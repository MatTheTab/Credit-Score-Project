```python
def fill_na_with_group_mode(df, groupby, column):
    col = df[column]
    mode_by_group = df.groupby(groupby)[column].transform(lambda x: x.mode()[0] if not x.mode().empty else np.NaN)
    result = col.fillna(mode_by_group)
    return result

credit_score["Occupation"] = credit_score["Occupation"].apply(lambda x: np.nan if x=='_______' else x)
print(credit_score["Occupation"].value_counts(dropna=False))

credit_score["Occupation"] = fill_na_with_group_mode(credit_score, 'Customer_ID', 'Occupation')
credit_score["Occupation"].value_counts(dropna=False)
```
